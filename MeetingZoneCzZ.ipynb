```python

import numpy as np
import matplotlib.pyplot as plt

# The Hunt for the Golden Frequency
grid_size = 100
frequencies = np.linspace(0.5, 1.2, 15) # Testing a range of 15 frequencies
stability_results = []

# Core Simulation Logic (Stripped for speed)
def run_test(omega_test):
    u = np.zeros((grid_size, grid_size))
    u_prev = np.zeros((grid_size, grid_size))
    V_gain = 0.15
    c_squared = 0.20 # Hydrogen medium acoustic velocity

    # Emitters
    emitter_x = 20
    x, y = np.arange(0, grid_size), np.arange(0, grid_size)
    X, Y = np.meshgrid(x, y)
    footprint = np.exp(-((X - emitter_x)**2) / (2 * 1.5**2)) * np.exp(-((Y - 50)**2) / (2 * 15.0**2))

    peak_force = 0
    force_history = []

    for t in range(300):
        laplacian = (np.roll(u,1,0)+np.roll(u,-1,0)+np.roll(u,1,1)+np.roll(u,-1,1)-4*u)
        u_next = 2*u - u_prev + c_squared * laplacian
        u_next += V_gain * np.sin(omega_test * t) * footprint
        u_next *= 0.994

        impact = np.abs(u[:, 90:]).mean()
        current_force = 0.5 * (impact**2)
        force_history.append(current_force)

        u_prev, u = u.copy(), u_next.copy()

    # Stability Score: Peak Force / Variance (We want high force, low dipping)
    stability_score = np.mean(force_history[-50:]) / (np.std(force_history[-50:]) + 1e-6)
    return stability_score, np.mean(force_history[-50:])

print("Starting Frequency Sweep... Let's find the peak!")
for omega in frequencies:
    score, avg_force = run_test(omega)
    stability_results.append((omega, score, avg_force))
    print(f"Freq: {omega:.2f} | Stability Score: {score:.4f} | Avg Force: {avg_force:.6f}")

# Find the winner
best_run = max(stability_results, key=lambda x: x[1])
print(f"\n--- WINNER FOUND ---")
print(f"Optimal Frequency: {best_run[0]:.2f}")
print(f"This is the frequency for your Patent and GitHub!")
```
